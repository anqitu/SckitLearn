# Module 5 Dimension Reduction

```python
import sklearn as sk
from sklearn import datasets
```

```python
# Load Iris dataset
iris = datasets.load_iris()
X,y = iris.data,iris.target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
```

```python
X_t= pca.fit_transform(X)
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_.cumsum()
```

```python
pca.singular_values_
```

## Exercise: PCA on Clustering

```python
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3)
cluster.fit(X)

cluster2 = KMeans(n_clusters=3)
cluster2.fit(X_t)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

plt.subplot(2,2,1)
plt.scatter(X[:,0],X[:,1],c=y)
plt.title('original')

plt.subplot(2,2,2)
plt.scatter(X[:,0],X[:,1],c=cluster.labels_)
plt.title('original')

plt.subplot(2,2,3)
plt.scatter(X_t[:,0],X_t[:,1],c=y)
plt.title('PCA')

plt.subplot(2,2,4)
plt.scatter(X_t[:,0],X_t[:,1],c=cluster2.labels_)
plt.title('PCA')
```

```python
# Split and Randomize the data
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=33)
```

```python
# KNN Classifier

from sklearn import neighbors
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```

```python
# Split and Randomize the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_t,y,test_size=0.25,random_state=33)
```

```python
# KNN Classifier

from sklearn import neighbors
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```
