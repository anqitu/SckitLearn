# Module 4 Clustering

## Load the Data

```python
%matplotlib inline
```

```python
from sklearn.datasets.samples_generator import make_blobs

centers = [[1,1],
           [1.5,1.5],
           [2,2]]
X,y = make_blobs(n_samples=100,
                 centers = centers,
                 cluster_std=0.2)
```

```python
# X
```

```python
# y
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])
plt.show()
```

## load the Model

```python
from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3,random_state=10)
```

## Train the Model

```python
cluster.fit(X)
```

## Evaluate the Model

```python
cluster.labels_
```

```python
cluster.cluster_centers_
```

```python
plt.subplot(1,2,1)
plt.scatter(X[:,0],X[:,1],c=y)
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,0],X[:,1],c=cluster.labels_)
plt.title("Prediction")
```

## KMeans Clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X,y = iris.data,iris.target

from sklearn.cluster import KMeans

cluster = KMeans(n_clusters=3,random_state=10)
cluster.fit(X)
```

```python
# cluster.labels_
```

```python
a = 1
b = 2
plt.subplot(1,2,1)
plt.scatter(X[:,a],X[:,b],c=y)
plt.xlabel(iris.feature_names[a])
plt.ylabel(iris.feature_names[b])
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,a],X[:,b],c=cluster.labels_)
plt.title("Prediction")
```

## MeanShift Clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X,y = iris.data,iris.target

from sklearn.cluster import MeanShift
cluster = MeanShift()
cluster.fit(X)

a = 0
b = 1
plt.subplot(1,2,1)
plt.scatter(X[:,a],X[:,b],c=y)
plt.xlabel(iris.feature_names[a])
plt.ylabel(iris.feature_names[b])
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,a],X[:,b],c=cluster.labels_)
plt.title("Prediction")
```

## Agglormerative Clustering

```python
from sklearn import datasets

iris = datasets.load_iris()
X,y = iris.data,iris.target

import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1])

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 3)
cluster.fit(X)

a = 1
b = 2
plt.subplot(1,2,1)
plt.scatter(X[:,a],X[:,b],c=y)
plt.xlabel(iris.feature_names[a])
plt.ylabel(iris.feature_names[b])
plt.title("Actual")
plt.subplot(1,2,2)
plt.scatter(X[:,a],X[:,b],c=cluster.labels_)
plt.title("Prediction")
```
